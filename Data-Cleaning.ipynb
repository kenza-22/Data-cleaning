```python
sprint = r"C:/Users/IMINFO/Desktop/projet/Back-end/sprints.json"
```

```python
df_sprint = pd.DataFrame(index=range(len(sprint)), columns= ['Id_sprint' ,'Nom_sprint', 'État' ,'Date_début' ,'Date_fin_prévue', 'Date_fin', 'Nom_Projet'])
df_sprint.head()
```

```python
df_sprint_list = []
with open(sprint, 'r') as f:
    data = json.load(f)

# Extraire les données des projets
if data:
    for item in data: 
        id_sprint = item.get('id', None)
        nom_sprint = item.get('name', None)
        etat = item.get('state', None) 
        debut = item.get('startDate', None)
        fin_prevu = item.get('endDate', None)
        fin = item.get('completeDate', None)
        projet = item.get('projectName', None)
        if etat == "active" and fin is None:
            fin = "Pas encore terminé"    
        if etat == "future" and fin is None and fin_prevu is None and debut is None:
            debut = "Pas encore défini" 
            fin_prevu = "Pas encore défini" 
            fin = "Pas encore défini" 
       
        # Ajouter une nouvelle ligne au DataFrame avec les données extraites
        row_data = {
            'Id_sprint': id_sprint,
            'Nom_sprint': nom_sprint,
            'État': etat,
            'Date_début': debut,
            'Date_fin_prévue': fin_prevu,
            'Fin': fin,
            'Nom_Projet': projet
        }
        df_sprint_list.append(row_data)

# Convertir la liste de dictionnaires en DataFrame
df_sprint = pd.DataFrame(df_sprint_list)

# Afficher la DataFrame
print(df_sprint)
```

```python
directory_path = r"C:\Users\IMINFO\Desktop\projet\Back-end\exports\issues"
```

```python
nom_fichiers=os.listdir(directory_path)
nom_fichiers
```

```python
df_issues = pd.DataFrame(index=range(len(nom_fichiers)), columns= ['Clé_Projet', 'Nom_Projet' ,'Id_sprint','Nom_sprint','Clé_Ticket', 'Id_Ticket', 'Priorité_Ticket', 'État_Ticket', 'Type_Ticket', 'Création', 'Mise à jour', 'Temps_Estimé', 'Temps_Consommé', 'Pourcentage_Progrès','Total_Progrès','Story_Point', 'Workratio'])
df_issues.head()
```

```python
import math 
df_issues_list = []
for idx, filename in enumerate(os.listdir(directory_path)):
    if filename.endswith('.json'):
        file_path = os.path.join(directory_path, filename)
        # Charger le contenu du fichier JSON
        with open(file_path, 'r') as f:
            data = json.load(f)
        # Extraire les valeurs depuis le fichier json
        if data:
            for item in data: 
                cle_ticket = item.get('key', None)
                timespent = item['fields'].get('timespent', None)
                creation = item['fields']['created']
                Maj = item['fields']['updated']
                if 'customfield_10020' in item['fields'] and item['fields']['customfield_10020']:
                    sprint = item['fields']['customfield_10020'][0]['name']
                else:
                    sprint = "Pas de sprint"
                if 'customfield_10020' in item['fields']:
                    customfield_10020_list = item['fields']['customfield_10020']
                if customfield_10020_list and len(customfield_10020_list) > 0:
                    first_item = customfield_10020_list[0]  # Accéder au premier élément de la liste
                    Id_sprint = first_item['id']
                else :
                    Id_sprint = 0
                story_point = item['fields']['customfield_10016']
                cle_projet = item['fields']['project']['key']
                nom_projet = item['fields']['project']['name']
                Temps_estime = item['fields'].get('timeestimate', None)
                if Temps_estime != 0 and Temps_estime is not None :
                    Temps_estimes = Temps_estime / 3600
                else :
                    Temps_estimes = 0
                if timespent is None or timespent == 0:
                    timespents = 0
                else :
                    timespents = timespent / 3600  
                priorite_ticket = item['fields']['priority']['name']
                Etat_ticket =  item['fields']['status']['name']
                Type_ticket =  item['fields']['issuetype']['name']
                workratio =  item['fields']['workratio']
                Progres = item['fields']['progress']['progress']
                total_progres = item['fields']['progress']['total']
                total_progress = total_progres / 3600
                if total_progress !=0:
                    if timespents == 0 :
                        pourcentage = (Temps_estimes / total_progress) * 100
                    elif timespents is not None and timespents !=0:
                        pourcentage = (timespents / total_progress) * 100
                else :
                    pourcentage = 0
                if pourcentage == 100.0 and timespents == 0 :
                    timespents = total_progress
                if story_point is None:
                    if timespent is not None and timespent != 0:
                        story_point = timespent / 7200
                    elif Temps_estime is not None and Temps_estime != 0:
                        story_point = Temps_estime / 7200
                # Ajouter une nouvelle ligne au Temps_estimeDataFrame avec les données extraites
                row_data = {
                    'Clé_Projet': cle_projet,
                    'Nom_Projet': nom_projet,
                    'Id_sprint': Id_sprint,
                    'Nom_sprint': sprint,
                    'Clé_Ticket': cle_ticket,
                    'État_Ticket': Etat_ticket,
                    'Type_Ticket': Type_ticket,
                    'Création': creation,
                    'Mise à jour': Maj,
                    'Temps_Estimé': Temps_estimes,
                    'Temps_Consommé': timespents,
                    'Story_Point': story_point,
                    'Total_Progrès': total_progress,
                    'Pourcentage_Progrès': pourcentage,
                }
                df_issues_list.append(pd.DataFrame([row_data]))
# Concatenate all DataFrame fragments into a single DataFrame
df_issues = pd.concat(df_issues_list, ignore_index=True)
```
