```python
import math 
df_issues_list = []
for idx, filename in enumerate(os.listdir(directory_path)):
    if filename.endswith('.json'):
        file_path = os.path.join(directory_path, filename)
        # Charger le contenu du fichier JSON
        with open(file_path, 'r') as f:
            data = json.load(f)
        # Extraire les valeurs depuis le fichier json
        if data:
            for item in data: 
                cle_ticket = item.get('key', None)
                timespent = item['fields'].get('timespent', None)
                creation = item['fields']['created']
                Maj = item['fields']['updated']
                if 'customfield_10020' in item['fields'] and item['fields']['customfield_10020']:
                    sprint = item['fields']['customfield_10020'][0]['name']
                else:
                    sprint = "Pas de sprint"
                if 'customfield_10020' in item['fields']:
                    customfield_10020_list = item['fields']['customfield_10020']
                if customfield_10020_list and len(customfield_10020_list) > 0:
                    first_item = customfield_10020_list[0]  # Accéder au premier élément de la liste
                    Id_sprint = first_item['id']
                else :
                    Id_sprint = 0
                story_point = item['fields']['customfield_10016']
                cle_projet = item['fields']['project']['key']
                nom_projet = item['fields']['project']['name']
                Temps_estime = item['fields'].get('timeestimate', None)
                priorite_ticket = item['fields']['priority']['name']
                Etat_ticket =  item['fields']['status']['name']
                Type_ticket =  item['fields']['issuetype']['name']
                workratio =  item['fields']['workratio']
                Progres = item['fields']['progress']['progress']
                total_progres = item['fields']['progress']['total']
                if total_progres !=0:
                    if timespent == 0 :
                        pourcentage = (Temps_estime / total_progres) * 100
                    elif timespent is not None and timespent !=0:
                        pourcentage = (timespent / total_progres) * 100
                else :
                    pourcentage = 0
                if pourcentage == 100.0 and timespent == 0 :
                    timespent = total_progres
                if story_point is None:
                    if timespent is not None and timespent != 0:
                        story_point = timespent / 7200
                    elif Temps_estime is not None and Temps_estime != 0:
                        story_point = Temps_estime / 7200
                # Ajouter une nouvelle ligne au Temps_estimeDataFrame avec les données extraites
                row_data = {
                    'Clé_Projet': cle_projet,
                    'Nom_Projet': nom_projet,
                    'Id_sprint': Id_sprint,
                    'Nom_sprint': sprint,
                    'Clé_Ticket': cle_ticket,
                    'État_Ticket': Etat_ticket,
                    'Type_Ticket': Type_ticket,
                    'Création': creation,
                    'Mise à jour': Maj,
                    'Temps_Estimé': Temps_estime,
                    'Temps_Consommé': timespent,
                    'Story_Point': story_point,
                    'Total_Progrès': total_progres,
                    'Pourcentage_Progrès': pourcentage,
                }
                df_issues_list.append(pd.DataFrame([row_data]))
# Concatenate all DataFrame fragments into a single DataFrame
df_issues = pd.concat(df_issues_list, ignore_index=True)
```

```python
groupes_projet = df_issues1.groupby("Clé_Projet")
```

```python
donnees_calculées = []

for projet, groupe in groupes_projet:
    # Calcul des données pour chaque projet
    tickets_estimes = groupe[groupe['Temps_Estimé'].notnull()].shape[0]
    tickets_non_estimes = groupe[groupe['Temps_Estimé'].isnull()].shape[0]
    charge_tickets_estimes = groupe[groupe['Temps_Estimé'].notnull()]['Temps_Estimé'].sum()
    tickets_par_etat = groupe.groupby('État_Ticket').size().to_dict()
    tickets_par_type = groupe.groupby('Type_Ticket').size().to_dict()
    charge_consomme = groupe[groupe['État_Ticket'].isin(['Terminé(e)', 'Fermée'])]['Temps_Consommé'].sum()
    charge_estime = groupe[groupe['État_Ticket'].isin(['Terminé(e)', 'Fermée'])]['Temps_Estimé'].sum()
    complexite_type = groupe.groupby('Type_Ticket')['Story_Point'].sum().to_dict()
    complexite_etat = groupe.groupby('État_Ticket')['Story_Point'].sum().to_dict()
    charge_etat = groupe[groupe['Temps_Estimé'].notnull()].groupby('État_Ticket')['Temps_Estimé'].sum().to_dict()
    charge_type = groupe[groupe['Temps_Estimé'].notnull()].groupby('Type_Ticket')['Temps_Estimé'].sum().to_dict()
    TE = groupe['Temps_Estimé'].sum()
    SP = groupe['Story_Point'].sum()
    if SP != 0:
        TEVsSP = TE / SP 
        Rapport = round(TEVsSP)
    else:
        Rapport = 0
    if charge_estime != 0:
        charge = charge_consomme / charge_estime
    else:
        charge = 0
    nbr_bug = groupe[groupe['Type_Ticket'].isin(['Bug', 'Bogue', 'Bug Sub Task'])].shape[0]
    nbr_bug_genere = groupe[(groupe['Type_Ticket'].isin(['Bug', 'Bogue', 'Bug Sub Task'])) & (groupe['État_Ticket'].isin(['Terminé(e)', 'Fermée']))].shape[0]
    nbr_SP_US = groupe[groupe['Type_Ticket'] == 'Story']['Story_Point'].sum()
    SP_done = groupe[groupe['État_Ticket'].isin(['Terminé(e)', 'Fermée'])]['Story_Point'].sum()
    Total_SP = groupe['Story_Point'].sum()
    if Total_SP != 0:
        SP_traite = SP_done / Total_SP
    else:
        SP_traite = 0
    if nbr_SP_US != 0:
        bug_genere_dev = nbr_bug_genere / nbr_SP_US
    else: 
        bug_genere_dev = 0    
    Charge_traitement_bug = groupe[groupe['Type_Ticket'].isin(['Bug', 'Bogue', 'Bug Sub Task'])]['Temps_Consommé'].sum()
    temps_passe_US = groupe[groupe['Type_Ticket'] == 'Story']['Temps_Consommé'].sum() 
    if temps_passe_US != 0:
        Qualite_fonctionnelle = Charge_traitement_bug / temps_passe_US
    else:
        Qualite_fonctionnelle = 0
    Charge_totale_estimation = groupe['Temps_Estimé'].sum()
    if Charge_totale_estimation != 0:
        Pourcentage_charge_traitement = (Charge_traitement_bug / Charge_totale_estimation) * 100
    else:
        Pourcentage_charge_traitement = 0
    Nbre_US_clôturés = groupe[(groupe['Type_Ticket'] == 'Story') & (groupe['État_Ticket'].isin(['Terminé(e)', 'Fermée']))].shape[0]
    Nbre_US_total = groupe[groupe['Type_Ticket'] == 'Story'].shape[0]
    if Nbre_US_total != 0:
        Ratio_bugs = nbr_bug / Nbre_US_total
    else:
        Ratio_bugs = 0
    if Nbre_US_total != 0:
        Avancement1 = (Nbre_US_clôturés / Nbre_US_total) * 100
    else:
        Avancement1 = 0
    Temps_estime = groupe['Temps_Estimé'].sum()
    Temps_consomme = groupe['Temps_Consommé'].sum()
    if Temps_estime != 0:
        Avancement2 = (Temps_consomme / Temps_estime) * 100
    else:
        Avancement2 = 0
    # Création du document avec les données calculées
    donnees_projet = {
        'Projet': projet,
        'Tickets_Estimés': int(tickets_estimes),  # Convertir en int
        'Tickets_Non_Estimés': int(tickets_non_estimes),  # Convertir en int
        'Charge_Tickets_Estimés': int(charge_tickets_estimes),  # Convertir en int
        'Tickets_par_Type': tickets_par_type,
        'Charge_par_Type': charge_type,
        'Tickets_par_État': tickets_par_etat,
        'Charge_par_État': charge_etat,
        'Complexité_par_Type': complexite_type,
        'Complexité_par_État': complexite_etat,
        'Charge': float(charge),  # Convertir en float
        'Rapport_SPTrait_TotalSP': float(SP_traite),  # Convertir en float
        'PE_VS_Charge': int(Rapport),  # Convertir en int
        'Nombre_Bugs': int(nbr_bug),  # Convertir en int
        'Charge_Traitement_Bug': float(Charge_traitement_bug),  # Convertir en float
        'Pourcentage_Charge_Traitement': float(Pourcentage_charge_traitement),  # Convertir en float
        'bug_genere_dev': float(bug_genere_dev),  # Convertir en float
        'Ratio_Bugs': float(Ratio_bugs),  # Convertir en float
        'Qualité_Fonctionnelle': float(Qualite_fonctionnelle),  # Convertir en float
        'Avancement_US': float(Avancement1),  # Convertir en float
        'Avancement_Temps': float(Avancement2)  # Convertir en float
    }
    
    # Insertion du document dans la collection MongoDB
    Tickets_collection.insert_one(donnees_projet)

```
